```python
#Applying formula and creating new column

print(ratings_mean_count_train['rating_counts'].sum(axis=0,skipna=True))
ratings_mean_count_train['count_ratings']=((ratings_mean_count_train['rating_counts']/ratings_mean_count_train['rating_counts'].sum(axis=0,skipna=True))*ratings_mean_count_train['Rating'])*1000
```

```python
#Sorting based on new column

ratings_mean_count_train.sort_values(by=['count_ratings','Rating'],ascending=False,inplace=True)
ratings_mean_count_train.head()
```

```python
popularity_recommendations=ratings_mean_count_train[ratings_mean_count_train['count_ratings']>5.0].head(5)

popularity_recommendations.sort_values(by=['Rating'],ascending=False,inplace=True)
print(popularity_recommendations)
```

```python
#reseting index so that it starts from 0

test_data=test_data.reset_index()
test_data=test_data.drop('index',axis=1)
test_data.head()
```

```python
# Use popularity based recommender model to make predictions
def recommend(user_id):     
    user_recommendations = popularity_recommendations 
          
    #Add user_id column for which the recommendations are being generated 
    user_recommendations['userID_idx'] = user_id 
      
    #Bring user_id column to the front 
    cols = user_recommendations.columns.tolist() 
    
    cols = cols[-1:] + cols[:-1] 
    user_recommendations = user_recommendations[cols] 
    user_recommendations['userId']=test_data.iloc[user_id].userId
    return user_recommendations 
```

# Collaborative filtering

```python
#Split data in train and test
train_data, test_data = train_test_split(df, test_size = 0.30, random_state=1)
```

```python
#Creating pivot table

#User-based Collaborative Filtering
# Matrix with row per 'user' and column per 'item' 
pivot_df = train_data.pivot(index = 'userId', columns ='productId', values = 'Rating').fillna(0)
pivot_df.shape
pivot_df.head()
```

```python
#info about pivot table
print('Shape of train pivot table: ', pivot_df.shape)
given_num_of_ratings = np.count_nonzero(pivot_df)
print('given_num_of_ratings = ', given_num_of_ratings)
possible_num_of_ratings = pivot_df.shape[0] * pivot_df.shape[1]
print('possible_num_of_ratings = ', possible_num_of_ratings)
density = (given_num_of_ratings/possible_num_of_ratings)
density *= 100
print ('density: {:4.2f}%'.format(density))
```

```python
#pivot table for test data

pivot_df_test = test_data.pivot(index = 'userId', columns ='productId', values = 'Rating').fillna(0)
pivot_df_test.shape
pivot_df_test.head()
```

```python
#info about pivot table
print('Shape of test pivot table: ', pivot_df_test.shape)

given_num_of_ratings = np.count_nonzero(pivot_df_test)
print('given_num_of_ratings = ', given_num_of_ratings)
possible_num_of_ratings = pivot_df_test.shape[0] * pivot_df_test.shape[1]
print('possible_num_of_ratings = ', possible_num_of_ratings)
density = (given_num_of_ratings/possible_num_of_ratings)
density *= 100
print ('density: {:4.2f}%'.format(density))
```

```python
#Creating a dummy column that will be used as an index later on

pivot_df['user_index'] = np.arange(0, pivot_df.shape[0], 1)
pivot_df.head()
```

```python
#Creating a dummy column that will be used as an index later on 

pivot_df_test['user_index'] = np.arange(0, pivot_df_test.shape[0], 1)
pivot_df_test.head()
```

```python
#Here index is being set as user_index for train data

pivot_df.set_index(['user_index'], inplace=True)

# Actual ratings given by users
pivot_df.head()
```

```python
#Here index is being set as user_index for test data

pivot_df_test.set_index(['user_index'], inplace=True)

# Actual ratings given by users
pivot_df_test.head()
```

```python
#shape of train pivot table
print(pivot_df.shape)
```

```python
#Collaborative filteing user user basedj8l

# Singular Value Decomposition
U, sigma, Vt = svds(pivot_df, k = 10)
# Construct diagonal array in SVD
sigma = np.diag(sigma)
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) 

# Predicted ratings 
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = pivot_df.columns)
preds_df.head()
```

```python
# Recommend the items with the highest predicted ratings

def recommend_items(userID, pivot_df_test, preds_df, num_recommendations):
      
    user_idx = userID-1 # index starts at 0
    
    # Get and sort the user's ratings
    sorted_user_ratings = pivot_df_test.iloc[user_idx].sort_values(ascending=False)
    #sorted_user_ratings
    sorted_user_predictions = preds_df.iloc[user_idx].sort_values(ascending=False)
    #sorted_user_predictions

    temp = pd.concat([sorted_user_ratings, sorted_user_predictions], axis=1)
    temp.index.name = 'Recommended Items'
    temp.columns = ['user_ratings', 'user_predictions']
    
    temp = temp.loc[temp.user_ratings == 0]   
    temp = temp.sort_values('user_predictions', ascending=False)
    print('\nBelow are the recommended items for user(user_id = {}):\n'.format(userID))
    print(temp.head(num_recommendations))
```

```python
#Enter 'userID' and 'num_recommendations' for the user in test set
userID =  4
num_recommendations = 5
recommend_items(userID, pivot_df_test, preds_df, num_recommendations)

#Note dont go beyond 1466 for userID
```

```python
# Actual ratings given by the users
final_ratings_matrix=pivot_df
final_ratings_matrix.head()
```

```python
# Average ACTUAL rating for each item
final_ratings_matrix.mean().head()
```

```python
# Predicted ratings 
preds_df.head()
```

```python
# Average PREDICTED rating for each item
preds_df.mean().head()
```

```python
#creating data frame to calculate rmse score for trainset
rmse_df = pd.concat([final_ratings_matrix.mean(), preds_df.mean()], axis=1)
rmse_df.columns = ['Avg_actual_ratings', 'Avg_predicted_ratings']
print(rmse_df.shape)
rmse_df['item_index'] = np.arange(0, rmse_df.shape[0], 1)
rmse_df.head()
```

```python
#RMSE for train set

RMSE = round((((rmse_df.Avg_actual_ratings - rmse_df.Avg_predicted_ratings) ** 2).mean() ** 0.5), 5)
print('\nRMSE SVD Model = {} \n'.format(RMSE))
```

```python
# Actual ratings given by the users
final_ratings_matrix_test=pivot_df_test
final_ratings_matrix_test.head()
```

```python
# Average ACTUAL rating for each item
final_ratings_matrix_test.mean().head()
```

```python
# Predicted ratings 
preds_df.head()
```

```python
# Average PREDICTED rating for each item
preds_df.mean().head()
```

```python
#creating data frame to calculate rmse score for test set
rmse_df = pd.concat([final_ratings_matrix_test.mean(), preds_df.mean()], axis=1)
rmse_df.columns = ['Avg_actual_ratings', 'Avg_predicted_ratings']
print(rmse_df.shape)
rmse_df['item_index'] = np.arange(0, rmse_df.shape[0], 1)
rmse_df.head()
```

```python
#RMSE for test set

RMSE = round((((rmse_df.Avg_actual_ratings - rmse_df.Avg_predicted_ratings) ** 2).mean() ** 0.5), 5)
print('\nRMSE SVD Model = {} \n'.format(RMSE))
```
